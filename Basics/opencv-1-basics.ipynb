# 3. Read, View and Export Images using OpenCV

- 1. Read in images into Python using OpenCV
- 2. Visualise images inside a Jupyter Notebook with Matplotlib
- 3. Write out transformed images

### Import Dependencies

````python
import cv2
import os
````



### Read and Render an Image
   

- Image path : `/kaggle/input/opencv-samples-python/lena.jpg`

````python
import matplotlib.pyplot as plt
````

````python
# Read an image
img_path = os.path.join('/kaggle/input/opencv-samples-python', 'lena.jpg')
img = cv2.imread(img_path)

# Show or Render image
plt.imshow(img)
````

### Filters

````python
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
````

````python
plt.imshow(recolor)
plt.show()
````

````python
plt.imshow(gray)
plt.show()
````

## Write out or Export Image

````python
cv2.imwrite('grayimage.jpg', gray)
````

````python
output_path = os.path.join('/kaggle/working', 'outputgray.jpg')
````
